```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import os
```

```python
filepath = "C:\\Users\\ginav\\Desktop\\Analysis Projects\\Project 1\\Project-1\\full_data.csv"
data = pd.read_csv(filepath)
#data.head()

complete_df = data.loc[~data.duplicated(subset=['County','State'])]
complete_df.describe()
```

## Education Visualizations

### Low Pop Density
#### Base Stats

```python
# create data frame
low_den_df = complete_df.loc[complete_df["Population Density (per m^2)"] < 250]
#low_den_df.head()
#low_den_df.columns
```

```python
xValues = low_den_df["Population Density (per m^2)"]
yLHS = low_den_df["Less Than A High School Diploma, 2017-21"]
yHS = low_den_df["High School Diploma Only, 2017-21"]
ySC = low_den_df["Some College Or Associate'S Degree, 2017-21"]
yBD = low_den_df["Bachelor'S Degree Or Higher, 2017-21"]
```

```python
plt.scatter(xValues, yLHS, edgecolors = 'black', alpha = .3)

plt.title("Population Density (per m^2) vs Number of people with \n Less than a High School Diploma")
plt.xlabel("Population Density Below 250 (per m^2)")
plt.ylabel("Number of People")

(slope, intercept, rvalue, pvalue, stderr) = linregress(xValues, yLHS)
regress_values = xValues * slope + intercept
line_eq = f"y = {round(slope,2)}x + {round(intercept,2)}\nR-value = {round(rvalue,2)}"

plt.plot(xValues,regress_values,"r-")
plt.annotate(line_eq,fontsize=12,color="red", xy=(150, 100000))
plt.plot(xValues, regress_values, "r-")

#plt.savefig("C:\\Users\\ginav\\Desktop\\Analysis Projects\\Project 1\\Project-1\\low-pop-less-than-hs-diploma.png", bbox_inches="tight")
```
