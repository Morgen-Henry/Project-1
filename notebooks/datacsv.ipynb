```python
import pandas as pd
```

```python
# import files
education = pd.read_excel("Education.xlsx")
poverty = pd.read_excel("PovertyEstimates.xlsx")
unemployment = pd.read_excel("Unemployment.xlsx")
```

```python
#change files to data frame
education_df = pd.DataFrame(education)
#education.head()
#print(list(education_df.columns))

poverty_df = pd.DataFrame(poverty)
#poverty_df.head()
#print(list(poverty_df.columns))

unemployment_df = pd.DataFrame(unemployment)
#unemployment_df.head()
#print(list(unemployment_df.columns))
```

```python
#clean data frame to keep only relevant data
education_columns = ["State", "Area name", "Less than a high school diploma, 2017-21", "High school diploma only, 2017-21", "Some college or associate's degree, 2017-21",
                    "Bachelor's degree or higher, 2017-21", "Percent of adults with less than a high school diploma, 2017-21",
                    "Percent of adults with a high school diploma only, 2017-21", "Percent of adults completing some college or associate's degree, 2017-21",
                    "Percent of adults with a bachelor's degree or higher, 2017-21"]
education_data = education_df[education_columns].copy()

education_data = education_data.drop(education_data[education_data["State"] == "PR"].index)
education_data.drop_duplicates(subset=["Area name", "State"], inplace=True)
education_data.dropna(inplace=True)
education_data = education_data[education_data['Area name'].str.contains('County')]

education_data = education_data.reset_index(drop=True)
education_data
```

```python
# clean data frame to keep only relevant data
poverty_columns = ["Stabr", "Area_name", "POVALL_2020", "PCTPOVALL_2020"]
poverty_data = poverty_df[poverty_columns].copy()

# rename columns to be more readable
poverty_data = poverty_data.rename(columns={"Stabr": "State", "Area_name": "Area name", "POVALL_2020": "Estimate of people of all ages in poverty 2020",
                                            "PCTPOVALL_2020": "Estimated percent of people of all ages in poverty 2020"})

poverty_data = poverty_data.drop(poverty_data[poverty_data["State"] == "PR"].index)
poverty_data.drop_duplicates(subset=["Area name", "State"], inplace=True)
poverty_data = poverty_data[poverty_data['Area name'].str.contains('County')]

poverty_data = poverty_data.reset_index(drop=True)
poverty_data
```

```python
#clean data frame to keep only relevant data
unemployment_columns = ["State", "Area_name", "Employed_2020", "Unemployed_2020", "Unemployment_rate_2020",
                        "Median_Household_Income_2020", "Med_HH_Income_Percent_of_State_Total_2020"]


unemployment_data = unemployment_df[unemployment_columns].copy()
#unemployment_data.head()

# rename columns to be more readable
unemployment_data = unemployment_data.rename(columns={"Area_name": "Area name", "Med_HH_Income_Percent_of_State_Total_2020": 
                                                    "County household median income as a percent of the State total median household income, 2020"})

unemployment_data = unemployment_data.drop(unemployment_data[unemployment_data["State"] == "PR"].index)
unemployment_data.drop_duplicates(subset=["Area name", "State"], inplace=True)
unemployment_data.dropna(inplace=True)
unemployment_data = unemployment_data[unemployment_data['Area name'].str.contains('County')]

unemployment_data = unemployment_data.reset_index(drop=True)
unemployment_data
```

```python
#clean data
education_data.columns = education_data.columns.str.title()
poverty_data.columns = poverty_data.columns.str.title()
unemployment_data.columns = unemployment_data.columns.str.title()
education_data.columns = education_data.columns.str.strip()
poverty_data.columns = poverty_data.columns.str.strip()
unemployment_data.columns = unemployment_data.columns.str.strip()


education_data = education_data.reset_index(drop=True)
poverty_data = poverty_data.reset_index(drop=True)
unemployment_data = unemployment_data.reset_index(drop=True)

all_data = pd.concat([education_data, poverty_data, unemployment_data], axis=1, join="outer").reindex(education_data.index)
all_data = all_data.loc[:, ~all_data.columns.duplicated()]
all_data = all_data.dropna()
all_data

```
