```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
filepath = "C:\\Users\\ginav\\Desktop\\Analysis Projects\\Project 1\\Project-1\\full_data.csv"
data = pd.read_csv(filepath)
data.head()
```

## Education Visualizations

```python
#function for regression
def linreg_function(x_values, y_values):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    return (regress_values, slope, intercept, rvalue)
```

### Low Pop Density
#### Base Stats

```python
# create data frame
low_den_df = data.loc[data["Population Density (per m^2)"] < 250]
#low_den_df.head()
low_den_df.columns
```

```python
xValues = low_den_df["Population Density (per m^2)"]
y1 = low_den_df["Less Than A High School Diploma, 2017-21"]
y2 = low_den_df["High School Diploma Only, 2017-21"]
y3 = low_den_df["Some College Or Associate'S Degree, 2017-21"]
y4 = low_den_df["Bachelor'S Degree Or Higher, 2017-21"]
```

```python
plt.scatter(xValues, y1)

plt.title("Population Density (m^2) vs Number of people with \n Less Than A High School Diploma")
plt.xlabel("Population Density (m^2)")
plt.ylabel("Number of People")

(slope, intercept, rvalue, pvalue, stderr) = linregress(xValues, y1)
regress_values = xValues * slope + intercept
line_eq = f"y = {round(slope,2)}x + {round(intercept,2)}\nR-value = {round(rvalue,2)}"

plt.plot(xValues,regress_values,"r-")
plt.annotate(line_eq,fontsize=12,color="red", xy=(150, 45))
plt.plot(xValues, regress_values, "r-")

plt.savefig("C:\\Users\\ginav\\Desktop\\Analysis Projects\\Project 1\\Project-1\\low-pop-less-than-hs-diploma")
```
